```python
import folium
```

```python
map = folium.Map(location=[35.157013, 129.051184], zoom_start=11)

for item in busan.index :
    lat = busan.loc[item , '위도']
    lang = busan.loc[item , '경도']
    
    item = item.strip()
    
    folium.CircleMarker([lat,lang],
                       radius = df1.loc[item, '재활용처리량']/5000,
                       popup = df1.loc[item, '재활용처리량'],
                       color= 'blue',
                       fill =True 
                       ).add_to(map)
    
    folium.CircleMarker([lat,lang],
                       radius = df1.loc[item, '음식물류발생량']/5000,
                       popup = df1.loc[item, '음식물류발생량'],
                       color= 'green',
                       fill =True 
                       ).add_to(map)

map
```
