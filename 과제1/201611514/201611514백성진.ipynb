### 201611514 백성진
#### 년도별 경남진주시 온도변화

```python
f = open('./201611514백성진.csv','r')
data=f.readlines()
f.close()
    
print(data)
```

```python
data_dic={}

for item in data :
    item.replace('\n','')
    line=item.split(',')
    data_dic[line[0]]=float(line[1])
print(data_dic)
```

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

y=data_dic.values()
```

```python
x=len(y)
x2=data_dic.keys()
plt.plot(x2,y,marker='o')
plt.xticks(rotation=90)
plt.title('연도별 평균온도변화(진주)',size=15)
plt.xlabel('년')
plt.ylabel('평균온도')
```
