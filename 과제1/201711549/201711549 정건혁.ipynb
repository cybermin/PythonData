# 201711549 정건혁 연도별 합계강수량과 평균상대습도 관의 관계

```python
f = open('./20190724151632.csv','r')
data = f.readlines()
f.close()

del data[0]

print(data)
```

```python
lst1 =[]
lst2 =[]
lst3 =[]

for i in data :
        i = i.replace("\n",'')
        item = i.split(',')
        
        lst1.append(item[0])
        lst2.append(float(item[1]))
        lst3.append(float(item[2]))

print(lst1)
print(lst2)
print(lst3)
```

```python
import matplotlib.pyplot as pit

y1 = lst2
y2 = lst3


x= range(len(lst1))
x2 = lst1

pit.plot(y1,'bo-')
pit.xticks(x,x2)
pit.show()
```

```python
import matplotlib.pyplot as pit

y1 = lst2
y2 = lst3

x= range(len(lst1))
x2 = lst1

pit.plot(y2,c='red',marker='v',linestyle='--')
pit.xticks(x,x2)
pit.show()
```

```python
import matplotlib.pyplot as pit

#한글 폰트 사용
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

y1 = lst2
y2 = lst3


x= range(len(lst1))
x2 = lst1

pit2.set_ylim(0,100)

pit.plot(x, y1, 'g-')
pit.bar(x, y2)

pit.xticks(x,x2,rotation=80)
pit.title('강수량과 습도',fontsize='20')
pit.xlabel('연도',fontsize='10')
pit.ylabel('강수량')

pit.show()
```

```python
fig, ax1 = pit.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-',marker='s')
ax2.plot(x, y2, 'b-',marker='o',linestyle='--')

ax1.set_xlabel('연도',fontsize=15)
ax1.set_ylabel('강수량', color='g',fontsize=10)
ax2.set_ylabel('습도량', color='b',fontsize=10)

pit.xticks(x,x2,rotation=80)
pit.title('강수량과 습도',fontsize='20')
ax2.set_ylim(50,80)

pit.show()
```
