## 201811524 윤민지
### 2016.01 - 2018.07 부산지역 평균상대습도

```python
f=open('부산습도.csv','r')
data=f.readlines()
f.close()

print(data)
```

```python
day = []
wet = []

for line in data :
    if line[:2] == '지점' : continue
    line = line.replace('\n','')
    item = line.split(',')
        
    day.append(item[1])
    wet.append(int(item[2]))
    
print(day)
print(wet)
```

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

plt.plot(wet)
plt.show()
```

```python
day_x=list(range(len(wet)))
plt.plot(wet,c='b',ls='--',lw=2,marker='*')
plt.plot(wet)
plt.xticks(day_x,day,rotation=90)  # rotation : 글자 회전
plt.title('부산 상대 습도',fontsize=20)  # 제목
plt.xlabel('일시')
plt.ylabel('평균 상대 습도')
plt.show()
```

```python
plt.scatter(day,wet)
plt.xticks(day_x,day,rotation=90)
plt.title('부산 상대 습도',fontsize=20)
plt.xlabel('일시')
plt.ylabel('평균 상대 습도')

plt.show()
```
