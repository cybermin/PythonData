#### 통영의 18년도 8월~19년도 6월 까지의 기온과 강수량 분석

```python
f = open('./PythonData-master/PythonData-master/02/기상청 프로젝트 데이터.csv')
data = f.read() 
f.close()

print(data) #통영의 18년도 8월 부터 19년도 6월 까지의 평균기온, 평균최고기온, 평균 최저기온, 일 최다 강수량 데이터
```

```python
type(data)
```

```python
data_list = data.split('\n')
print(data_list)
data_list.pop()
print(data_list)
del data_list[-1]
del data_list[0]
print(data_list)
```

```python
data_평균기온 = {}
data_평균최고기온 = {}
data_평균최저기온 = {}
data_일최다강수량 = {}
for i in data_list :
    line = i.split(',')
    data_평균기온[line[1]] = float(line[2])
    data_평균최고기온[line[1]] = float(line[3])
    data_평균최저기온[line[1]] = float(line[4])
    data_일최다강수량[line[1]] = float(line[5])

print(data_평균기온)
print(data_평균최고기온)
print(data_평균최저기온)
print(data_일최다강수량)
```

```python
import matplotlib.pyplot as plt
#한글 폰트사용
from matplotlib import font_manager, rc
font_name =font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
y1 = data_평균기온.values()
y2 = data_평균최고기온.values()
y3 = data_평균최저기온.values()
y4 = data_일최다강수량.values()

x1 = range(len(y1))
x2 = range(len(y2))
x3 = range(len(y3))
x4 = range(len(y4))


x0 = data_평균기온.keys()

plt.plot(x1,y1,c='g',label='평균온도',lw=2,ls=':')
plt.plot(x2,y2,c='r',label='평균최대온도',lw=4,ls=':')
plt.plot(x3,y3,c='y',label='평균최저온도',lw=4,ls=':')
plt.plot(x4,y4,c='b',label='최다강수량',lw=4,ls='--')
plt.xticks(x1, x0, rotation=90)
plt.xlabel('년월')
plt.ylabel('온도')
plt.title('월별 통영의 온도변화')
plt.legend(loc='upper left')
plt.show()
```

```python
fig, ax1 = plt.subplots()

x = data_평균기온.keys()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', ls = ':')
ax1.plot(x, y2, 'r-')
ax1.plot(x, y3, 'y-')
ax2.plot(x, y4, 'b-', ls = '--')

plt.xticks(x1, x0, rotation=90)
ax1.set_xlabel('년월')
ax1.set_ylabel('온도', color='g')
ax2.set_ylabel('강수량', color='b')

plt.show() 
#18년도 8월~19년도 6월 사이의 통영의 최고 기온을 찍은 기간은 18년도 8월이고, 최저 기온을 찍은 기간은 19년도 1월이다.
#위의 결과를 강수량과 비교를 해보면 우리는 무더운 여름에 비가 많이 온다고 생각을 하지만 밑의 그래프를 보면
#오히려 가장 더운 날이 아닌 19년도 6월에 가장 많은 강수량이 찍혀 있음을 알 수 있다.
#18년도의 경우 또한 가장 더웠던 8월이 아닌 9월에 강수량이 많은 것 을 알 수 있다.
#이는 생각 해보면 가장 덥다는 말은 강한 고기압이 발생하여 구름을 만들 수 없는 환경을 조성하기 때문에 비가 올 확률이 낮아짐을 알 수있다.
```
