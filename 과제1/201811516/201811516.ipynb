#### 경주 월별 평균기온 변화

```python
f=open('201811516(경주시 월별 기온).csv','r')
data=f.readlines()
f.close()
print(data)
```

```python
data_tem = {}

for item in data :
    item = item.replace('\n','')
    line = item.split(',')
    
    data_tem[line[0]] = float(line[1])

print(data_tem)
```

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

y = data_tem.values()
```

```python
plt.plot(y)
plt.show()
```

```python
x = range(len(y))
x2 = data_tem.keys()

plt.plot(y)


plt.title('경주 월별 기온',fontsize=10)
plt.xlabel('월')
plt.ylabel('기온')
plt.show()
```
