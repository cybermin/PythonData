### 부산 월별 최대풍속

```python
ex=open('./20190724153939.csv')
data=ex.read()
ex.close()
```

```python
data_list=data.split('\n')
data_list.remove('지점,일시,최대풍속(m/s)')
data_list.pop()
```

```python
data_dic={}
for item in data_list :
    item = item.replace('\n','')
    line = item.split(',')
    data_dic[line[1]]=float(line[2])
```

```python
import matplotlib.pyplot as data 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

y=data_dic.values()
```

```python
data.plot(y)
x=range(len(y))
x2=data_dic.keys()
data.xticks(x,x2,rotation=90)
data.title('부산 월별 최대풍속')
data.show()
```

```python
x=range(len(y))
x2=data_dic.keys()

data.bar(x,y)
data.xticks(x,x2,rotation=90)
data.title('부산 월별 최대풍속')
data.show()
```
