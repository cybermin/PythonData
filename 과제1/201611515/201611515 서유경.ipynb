#### 부산시의 6월 평균 기온

```python
f=open("data.csv",'r')
data=f.readlines()
f.close()

print(data)
```

```python
data.pop(0)
print(data)
```

```python
x = []
y = []
etc=[]

for line in data :
    line=line.replace('\n','')
    item = line.split(',')
    etc.append(item[0])
    x.append(item[1])
    y.append(float(item[2]))

print(x)
print(y)
```

```python
etc = ['부산' for i in etc]
print(etc)
```

```python
import matplotlib.pyplot as plt
plt.plot(y)
plt.show()
```

```python
dic=dict(zip(x,y))
print(dic)
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
```

```python
import matplotlib.pyplot as plt

x=list(range(len(dic)))
y=list(dic.values())
plt.title('부산시 6월 평균 기온 변화',fontsize = 15)
plt.plot(y)
plt.xticks(x, list(dic.keys()),rotation=90)
plt.show()

```
