#### 월별 경남지역(창원_155지점) 최대순간풍속

```python
def rcsv(filename) :
    lst1 = []
    lst2 = []
    lst3 = []
    
    f = open(filename,'r')
    data=f.readlines()
    f.close()
    
    for line in data :
        line = line.replace('\n','')
        item = line.split(',')
        lst1.append(item[0])
        lst2.append(item[1])
        lst3.append(item[2])
        
    return lst1, lst2, lst3

area = []
time = []
wind = []

area, time, wind = rcsv("./최대순간풍속_155.csv")

del area[0]
del time[0]
del wind[0]

#print(time)
#print(wind)


import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)



x = range(len(time))

plt.plot(wind, c='r', lw=1.5, ls=':', marker='x')
plt.title('월 별 최대 순간 풍속 추이', fontsize=20)
plt.xticks(x, time, rotation=90,fontsize=5)
plt.xlabel('시간', fontsize=15)
plt.ylim(0,30)
plt.show()
```
