```python
#파일불러오는 함수
def rcsv2(filename):
    lst1=[]
    lst2=[]
    lst3=[]
    
    f = open(filename, 'r')
    data=f.readlines()
    f.closed
    
    for line in data :
        line=line.replace('\n','')
        item = line.split(',')
    
        lst1.append(item[0])
        lst2.append(item[1]) 
        lst3.append(item[2])
    return lst1, lst2, lst3
```

```python
날짜, 평균최고기온, 평균최저기온 =rcsv2('청주평균기온.csv')
```

```python
del 날짜[0]
del 평균최고기온[0]
del 평균최저기온[0]
```

```python
print(날짜)
평균최고기온 = [float(i) for i in 평균최고기온]
평균최저기온 = [float(i) for i in 평균최저기온]
print(평균최고기온)
print(평균최저기온)
```

```python
x = list(range(len(날짜)))

plt.plot(평균최고기온,label='평균최고기온',marker="*",c="magenta")
plt.plot(평균최저기온,label='평균최저기온',marker="o")
plt.xticks(x,날짜,rotation=90)  #x축에 이름 넣어주기, rotation하면 글자 회전 가능
plt.title('청주 평균기온',fontsize=20) #title이름 넣기
plt.xlabel('날짜',fontsize=12)#fontsize=> 폰트크기
plt.ylabel('기온',fontsize=12)
plt.ylim((-20,40))
plt.legend()
plt.show()
```
