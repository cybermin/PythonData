```python
f = open('20190724151623.csv','r')
data = f.readlines()
f.close()

print(data)
```

```python
lst1=[]
lst2=[]

for i in data :
    j=i.replace('\n','')
    k=j.split(',')
    
    lst1.append(k[0])
    lst2.append(int(k[1]))
    
print(lst1)
print(lst2)
```

```python
import matplotlib.pyplot as plt

plt.plot(lst2)
plt.show()
```

```python
lst=list(range(len(lst1)))
plt.plot(lst2)
plt.xticks(lst,lst1,rotation=90)
plt.show()
```
