```python
import pandas as pd
```

```python
so1 = pd.read_csv('./01.csv', engine='python')
```

```python
so1.head()
```

```python
data = so1[so1['시도명'] == '부산광역시']
```

```python
data = data[['상권업종대분류명','시군구명','법정동명','경도','위도']]
```

```python
data.head()
```

```python
data['상권업종대분류명'].value_counts()
```

```python
data = data.set_index('상권업종대분류명')
```

```python
data.head()
```

```python
data1 = data[data.index == "소매"]
data2 = data[data.index == "음식"]
data3 = data[data.index == "생활서비스"]
data4 = data[data.index == "학문/교육"]
data5 = data[data.index == "부동산"]
data6 = data[data.index == "관광/여가/오락"]
data7 = data[data.index == "숙박"]
data8 = data[data.index == "스포츠"]
```
