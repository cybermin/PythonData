# 직접 선정한 데이터로 분석 진행하기

```python
import pandas as pd
data = pd.read_csv('월별교통사고.csv', engine='python')
data.head()
##변수 설명##
#시도 : 사고가 발생한 시, 도
#시군구 : 사고가 발생한 군, 구 
#월 : 사고가 발생한 월
#발생건수 : 사고가 발생한 횟수
#사망자수 : 사고로 인한 사망자수
#부상자수 : 사고로 인한 부상자수
#중상 : 부상자 중 중상을 입은 사람의 수
#경상 : 부상자 중 경상을 입은 사람의 수
#부상신고 : 부상자 중 부상신고를 입은 사람의 수
```

```python
#부산의 데이터로 축약하기
data2 = data[data['시도']=='부산']
data2
```

```python
#부산대학교가 속해져 있는 구인 금정구 데이터 뽑아보기
data3 = data2[data2['시군구']=='금정구']
data3
```

```python
data4 = data3[['월','발생건수']]
data4
```

```python
data4 = data4.set_index('월')
data4
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name =font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

x1 = range(len(data4['발생건수']))
x0 = list(data4.index.values)

data4.plot(figsize=(10,6))
plt.title('월별 금정구 교통사고발생건수',size=20)
plt.xticks(x1, x0)
plt.show()
#11월이 가장 교통사고가 빈번하게 발생하는 월로 볼 수 있다.
#반면에 1월이 가장 교통사고가 발생하지 않는 월로 볼 수 있다.
```

```python
#월별 부산 구의 발생건수 비교
data6 = data2[['시군구','월','발생건수']]
data6 = data6.set_index('월')
data6
```

```python
Y1 = data6[data6['시군구']=='중구']['발생건수']
Y2 = data6[data6['시군구']=='서구']['발생건수']
Y3 = data6[data6['시군구']=='동구']['발생건수']
Y4 = data6[data6['시군구']=='영도구']['발생건수']
Y5 = data6[data6['시군구']=='부산진구']['발생건수']
Y6 = data6[data6['시군구']=='동래구']['발생건수']
Y7 = data6[data6['시군구']=='남구']['발생건수']
Y8 = data6[data6['시군구']=='북구']['발생건수']
Y9 = data6[data6['시군구']=='해운대구']['발생건수']
Y10 = data6[data6['시군구']=='사하구']['발생건수']
Y11 = data6[data6['시군구']=='금정구']['발생건수']
Y12 = data6[data6['시군구']=='강서구']['발생건수']
Y13 = data6[data6['시군구']=='연제구']['발생건수']
Y14 = data6[data6['시군구']=='수영구']['발생건수']
Y15 = data6[data6['시군구']=='사상구']['발생건수']
Y16 = data6[data6['시군구']=='기장군']['발생건수']

x = range(len(Y1))
x1 = list(data6.index.values)
```

```python
plt.plot(Y1,c='k',label='중구')
plt.plot(Y2,c='grey',label='서구')
plt.plot(Y3,c='lightgray',label='동구')
plt.plot(Y4,c='lightcoral',label='영도구')
plt.plot(Y5,c='maroon',label='부산진구')
plt.plot(Y6,c='coral',label='동래구')
plt.plot(Y7,c='orange',label='남구')
plt.plot(Y8,c='olive',label='북구')
plt.plot(Y9,c='yellowgreen',label='해운대구')
plt.plot(Y10,c='lawngreen',label='사하구')
plt.plot(Y11,c='lightgreen',label='금정구')
plt.plot(Y12,c='g',label='강서구')
plt.plot(Y13,c='c',label='연제구')
plt.plot(Y14,c='skyblue',label='수영구')
plt.plot(Y15,c='b',label='사상구')
plt.plot(Y16,c='navy',label='기장군')
plt.xticks(x, x1)
plt.title('월별 부산 구의 교통사고발생건수')
plt.xlabel('월')
plt.ylabel('사고발생건수')
plt.show()
```
