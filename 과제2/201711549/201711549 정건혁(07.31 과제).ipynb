```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
data = pd.read_csv('./downloads/data/이혼.csv',engine='python')
```

```python
data= data.T
```

```python
data=data.reset_index()
```

```python
data
```

```python
data = data.drop(0)
```

```python
data.head()
```

```python
data.columns = ['년수','항목','전국','부산']
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.set_index('년수')
data
```

```python
data = data.rename(index ={'2018 p)':'2018',
                  '2018 p).1':'2018.1',
                  '2018 p).2':'2018.2',
                  '2018 p).3':'2018.3',
                 '2018 p).4':'2018.4',
                  '2018 p).5':'2018.5'}
                  )
```

```python
data =data.reset_index()
```

```python
data['년수'] = pd.to_numeric(data['년수'])
```

```python
data.head()
```

```python
data1 = data[data['항목']=='출생아수(명)']
data1 
```

```python
data1=data1.drop('항목',axis=1)
```

```python
data1 = data1.set_index()
```

```python
data2 = data[data['항목']=='혼인건수(건)']
data2
```

```python
data2=data2.drop('항목',axis=1)
```

```python
data2
```

```python
data2['년수']=round(data2['년수'])
```

```python
data2
```

```python
data3 = pd.merge(data1,data2, on='년수',how='inner')
```

```python
data3 =  data3.set_index('년수')
```

```python
data3
```

```python
data4 = data3.drop('부산_x',axis=1)
```

```python
data4
```

```python
data4 = data4.drop('부산_y',axis=1)
```

```python
data4.columns = ['출생아수(전국)','혼인건수(전국)']
```

```python
data4['출생아수(전국)'] = pd.to_numeric(data4['출생아수(전국)'])
data4['혼인건수(전국)'] = pd.to_numeric(data4['혼인건수(전국)'])
```

```python
data4.plot()
plt.title('혼인건수,출생아수 상관그래프',size=20)
plt.show()
```

```python
data5 = data3.drop('전국_x',axis=1)
```

```python
data5 = data5.drop('전국_y',axis=1)
```

```python
data5
```

```python
data5.columns = ['출생아수(부산)','혼인건수(부산)']
```

```python
data5['출생아수(부산)'] = pd.to_numeric(data5['출생아수(부산)'])
data5['혼인건수(부산)'] = pd.to_numeric(data5['혼인건수(부산)'])
```

```python
data5
```

```python
data5.plot()
plt.title('혼인건수,출생아수 상관그래프',size=20)
plt.show()
```

```python
data6 = pd.merge(data4,data5,on='년수',how='inner')
```

```python
data6
```

```python
data6.plot(kind='bar')
plt.title('혼인건수,출생아수 상관그래프',size=20)
plt.show()
```

```python
y1 = data4
y2 = data5

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot( y1, 'g-',marker='s')
ax2.plot( y2, 'b-',marker='o',linestyle='--')

ax1.set_ylabel('전국', color='g',fontsize=20)
ax2.set_ylabel('부산', color='b',fontsize=20)

ax2.set_ylim(0,40000)

plt.show()

# 선 별로 색깔을 달리 해주려고 했으나 실패했음.
```
