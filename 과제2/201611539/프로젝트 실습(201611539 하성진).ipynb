```python
import pandas as pd
```

```python
lib = pd.read_csv('전국도서관표준데이터.csv', engine = 'python')
```

```python
lib
```

```python
lib = lib.rename(columns={'자료수(도서)':'도서','자료수(연속간행물)':'연속간행물','자료수(비도서)':'비도서'})
```

```python
lib2 = lib.groupby('시도명').mean()
lib2
```

```python
import folium

map = folium.Map(location=[35.166804, 129.083479], zoom_start=6)

for item in lib2.index:
    lat = lib2.loc[item, '위도']
    long = lib2.loc[item, '경도']

    folium.CircleMarker([lat,long],
                       radius=lib2.loc[item, '도서']/8000,
                       popup=lib2.loc[item,'도서'],
                       color='crimson',fill=True).add_to(map)
    
map
```
