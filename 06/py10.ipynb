```python
import folium
map = folium.Map(location=[35.157013, 129.051184], zoom_start=11)

for item in df2.index :
    lat = df2.loc[item , '위도']
    lang = df2.loc[item , '경도']
    
    folium.CircleMarker([lat,lang],
                       radius = df2.loc[item, '가스'],
                       popup = df2.loc[item, '가스'],
                       color= 'blue',
                       fill =True 
                       ).add_to(map)
    folium.CircleMarker([lat,lang],
                       radius = df2.loc[item, '전기'],
                       popup = df2.loc[item, '전기'],
                       color= 'green',
                       fill =True 
                       ).add_to(map)
map    
```
