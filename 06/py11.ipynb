```python
df['일시'] = pd.to_datetime(df['일시'])
```

```python
df['년도'] = df['일시'].dt.year
```

```python
df['월'] = df['일시'].dt.month
```

```python
df8 = df[df['월']==8]
```

```python
df8g = df8.groupby(['년도'])['평균','최고','최저'].mean()
```
